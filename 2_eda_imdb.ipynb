### 7. Renombra las columnas que tengan espacios

### 8. Utiliza tus nuevas columnas sin espacios :)

```python
df.Runtime_Minutes.head()      # Notación punto
df["Runtime_Minutes"].head()   # Notación corchetes
# Películas con más de 120 minutos
df[df.Runtime_Minutes > 120]
# Películas con ingresos mayores a 100 millones
df[df.Revenue_Millions > 100]
df.Runtime_Minutes.mean()
df.Revenue_Millions.sum()
df.assign(
    Hours = df.Runtime_Minutes / 60
).query("Hours > 2").head()
plt.hist(df.Runtime_Minutes, bins=10)
plt.xlabel("Runtime (minutes)")
plt.ylabel("Frequency")
plt.title("Distribución de duración de películas")
plt.show()

```

### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con essos valores

```python
df.Rating.unique()
df.Rating.value_counts()
rating_counts = df.Rating.value_counts().reset_index()
rating_counts.columns = ["Rating", "Frecuencia"]
rating_counts
```

### 22. Ordena el nuevo dataframe por la variable `rating`

```python
rating_counts.sort_values(by="Rating", inplace=True)
rating_counts
```

### 23. Crea una gráfica de barras con este nuevo dataframe ordenado

```python
plt.bar(rating_counts["Rating"], rating_counts["Frecuencia"], edgecolor="black")
plt.xlabel("Rating")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de Ratings (ordenados)")
plt.xticks(rotation=45)
plt.show()
```

### 24. Crea la matriz de correlación del dataframe de películas

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])
corr_matrix = numeric_df.corr()
```

### 25. Grafica la matriz de correlación utilizando `matshow()`

```python

plt.figure(figsize=(10,6))
plt.matshow(corr_matrix, cmap="coolwarm", fignum=1)
plt.colorbar()
plt.title("Matriz de correlación (matshow)", pad=20)

# Etiquetas de los ejes
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

plt.show()
```
